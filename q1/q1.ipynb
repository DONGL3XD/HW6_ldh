```python
import csv
import matplotlib.pyplot as plt
#,,(),(),()

def main() :
    seoul = open('seoul.csv', 'r', encoding='cp949')
    sdata = csv.reader(seoul, delimiter = ',')
    daejeon = open('daejeon.csv', 'r', encoding='cp949')
    ddata = csv.reader(daejeon, delimiter = ',')
    busan = open('busan.csv', 'r', encoding='cp949')
    bdata = csv.reader(busan, delimiter = ',')
    jeju = open('jeju.csv', 'r', encoding='cp949')
    jdata = csv.reader(jeju, delimiter = ',')
    overall = open('korea.csv', 'r', encoding='cp949')
    tdata = csv.reader(overall, delimiter = ',')

    stemp = []
    dtemp = []
    btemp = []
    jtemp = []
    ttemp = []

    for row in sdata :
        stemp.append(float(row[2]))
    for row in ddata :
        dtemp.append(float(row[2]))
    for row in bdata :
        btemp.append(float(row[2]))
    for row in jdata :
        jtemp.append(float(row[2]))
    for row in tdata :
        ttemp.append(float(row[2]))

    plt.title("Average Temp - Yearly")
    plt.ylabel('Temp')
    plt.plot(stemp, color='red', label='seoul')
    plt.plot(dtemp, color='orange', label='daejeon')
    plt.plot(btemp, color='yellow', label='busan')
    plt.plot(jtemp, color='green', label='jeju')
    plt.plot(ttemp, color='blue', label='overall')
    plt.show()


    for i in range(12) :
        stemp[i]  = round(stemp[i] - ttemp[i], 2)
        dtemp[i]  = round(dtemp[i] - ttemp[i], 2)
        btemp[i]  = round(btemp[i] - ttemp[i], 2)
        jtemp[i]  = round(jtemp[i] - ttemp[i], 2)

    print("seoul : ", stemp)
    print("Daejeon : ", dtemp)
    print("Busan : ", btemp)
    print("Jeju : ", jtemp)

if __name__=="__main__" :
    main()
```
